# NLP Assignment - Word Embeddings and Representation Learning
## Overview
This notebook covers the progression of word representations using Frequency-based, Prediction-based, and Contextualized embeddings.

**Dataset**: Twitter Sentiment (Roll No 23F-0031 is Odd). We use a small subset of tweets to ensure the notebook runs very fast.

### Setup: Dataset Loading
We will load standard tools and define our dataset. It is an array of strings (`X`) and their sentiments (`y`).

```python
import warnings
warnings.filterwarnings('ignore')

from datasets import load_dataset
import pandas as pd
import numpy as np

# Load a small subset of Twitter sentiment data for speed
print("Loading Twitter Sentiment dataset...")
try:
    # "tweet_eval" dataset with "sentiment" subset
    dataset = load_dataset("tweet_eval", "sentiment")
    train_df = pd.DataFrame(dataset['train']).sample(1000, random_state=42)
    test_df = pd.DataFrame(dataset['test']).sample(200, random_state=42)
except Exception as e:
    print("Could not load from huggingface, using dummy twitter data for demonstration.")
    train_df = pd.DataFrame({"text": ["I love this!", "I hate it.", "this is okay"] * 334, "label": [2, 0, 1] * 334})
    test_df = pd.DataFrame({"text": ["great app", "terrible UI"] * 100, "label": [2, 0] * 100})

X_train = train_df['text'].tolist()
y_train = train_df['label'].tolist()
X_test = test_df['text'].tolist()
y_test = test_df['label'].tolist()

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
```

## Module 1: Frequency-Based Embeddings
**Concepts:** Bag of Words (BoW) & TF-IDF
**Method:** We use simple Scikit-Learn vectorizers.

**Steps Done Behind the Scenes by CountVectorizer:** Lowercasing, stopword removal, and basic tokenization.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

print("\n--- Training Bag of Words (BoW) ---")
# Limit max_features to 5000 so it's not too huge
bow_vectorizer = CountVectorizer(stop_words='english', lowercase=True, max_features=5000)
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

clf_bow = LogisticRegression(max_iter=1000)
clf_bow.fit(X_train_bow, y_train)
y_pred_bow = clf_bow.predict(X_test_bow)

acc_bow = accuracy_score(y_test, y_pred_bow)
print(f"BoW Accuracy: {acc_bow * 100:.2f}%")

print("\n--- Training TF-IDF ---")
tfidf_vectorizer = TfidfVectorizer(stop_words='english', lowercase=True, max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

clf_tfidf = LogisticRegression(max_iter=1000)
clf_tfidf.fit(X_train_tfidf, y_train)
y_pred_tfidf = clf_tfidf.predict(X_test_tfidf)

acc_tfidf = accuracy_score(y_test, y_pred_tfidf)
print(f"TF-IDF Accuracy: {acc_tfidf * 100:.2f}%")
```

### Observation (Module 1)
* **Limitations:** Frequency-based methods ignore context. Two sentences with different meanings but similar overlapping words will have highly similar vectors.
* **Context Discarded:** Standard BoW often removes words like "not" due to stopword filtering, failing to understand simple negations like "not good". Matrices are also extremely large and sparse.

## Module 2: Prediction-Based Embeddings
**Concepts:** Word2Vec (GloVe used for simplicity)
**Method:** We load small 25-dimensional pre-built word vectors from Gensim. A review is then represented as the pure average of its words.

```python
import gensim.downloader as api
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

print("\nLoading pre-built GloVe embeddings (prediction-based)...")
# Using glove-twitter-25 because it's very small and fast to download
word_vectors = api.load("glove-twitter-25")
dim = 25

def document_vector(doc):
    words = doc.lower().split() 
    vecs = [word_vectors[w] for w in words if w in word_vectors]
    if len(vecs) > 0:
        return np.mean(vecs, axis=0) # Averaging the word vectors
    return np.zeros(dim)

X_train_w2v = np.array([document_vector(text) for text in X_train])
X_test_w2v = np.array([document_vector(text) for text in X_test])

print("Training Simple Neural Network with Word embeddings...")
# Simple Feedforward Neural Network
clf_nn = MLPClassifier(hidden_layer_sizes=(50,), max_iter=500, random_state=42)
clf_nn.fit(X_train_w2v, y_train)
y_pred_w2v = clf_nn.predict(X_test_w2v)

acc_w2v = accuracy_score(y_test, y_pred_w2v)
print(f"Prediction-Based Embeddings Accuracy: {acc_w2v * 100:.2f}%")

print("\nDeploying PCA to visualize word embeddings...")
words_to_plot = ["good", "great", "bad", "terrible", "movie", "film", "director", "actor"]
# Extract vectors only for words that actually exist in the vocab
valid_words = [w for w in words_to_plot if w in word_vectors]
vectors_to_plot = np.array([word_vectors[w] for w in valid_words])

# Apply PCA down to 2 dimensions for easy graph plotting
pca = PCA(n_components=2)
result = pca.fit_transform(vectors_to_plot)

plt.figure(figsize=(8,6))
plt.scatter(result[:, 0], result[:, 1])
for i, word in enumerate(valid_words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]), xytext=(5, 5), textcoords='offset points')
plt.title("PCA Visualization of Word Embeddings")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
```

### Observation (Module 2)
* **Semantic Similarity:** As seen on the plot, logically similar words (e.g., 'movie' and 'film') are clustered closer mathematically.
* **Performance Difference:** Often, pre-built embeddings perform comparably or better than BoW, but with vastly fewer dimensions (e.g., 25 dimensions vs 5000+ sparse dimensions).

## Module 3: Contextualized Word Embeddings
**Concepts:** BERT-based extraction
**Method:** We use the HuggingFace `pipeline` to get embeddings. 

```python
from transformers import pipeline
import torch

print("\nLoading Contextualized Model (DistilBERT)...")
# feature-extraction gives the raw embeddings
extract = pipeline('feature-extraction', model='distilbert-base-uncased', device=-1)

# ---- Polysemy Check ----
print("\n---- Checking Polysemy ----")
# 'Bank' has two different meanings in these sentences.
sentence1 = "I sat by the river bank."
sentence2 = "I deposited money at the bank."

out1 = extract(sentence1)
out2 = extract(sentence2)

# Approximate indices of the word 'bank':
# "I", "sat", "by", "the", "river", "bank", "." --> Index 6
# "I", "deposited", "money", "at", "the", "bank", "." --> Index 6
vec1 = np.array(out1[0][6])
vec2 = np.array(out2[0][6])

# Cosine similarity tells us how matching two vectors are
similarity = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
print(f"Cosine Similarity between 'bank'(river) and 'bank'(money) under BERT: {similarity:.4f}")
print("If we used Word2Vec from Module 2, the similarity would just be 1.0 because the word representation is static.")

# ---- Classifier with Contextual Embeddings ----
# Limit sizes aggressively here because CPU BERT generation takes very long without a GPU
subset_train = 200
subset_test = 50

print(f"\nComputing Contextual Embeddings for a small subset ({subset_train} Train / {subset_test} Test) to save processing time...")
def get_sentence_embedding(text):
    # Take the [CLS] token (0th token) as whole sentence representation
    out = extract(text, truncation=True, max_length=128)
    return np.array(out[0][0])

X_train_bert = np.array([get_sentence_embedding(t) for t in X_train[:subset_train]])
X_test_bert = np.array([get_sentence_embedding(t) for t in X_test[:subset_test]])
y_train_bert = y_train[:subset_train]
y_test_bert = y_test[:subset_test]

# Simple Logistic Regression classifier
clf_bert = LogisticRegression(max_iter=500)
clf_bert.fit(X_train_bert, y_train_bert)
y_pred_bert = clf_bert.predict(X_test_bert)

acc_bert = accuracy_score(y_test_bert, y_pred_bert)
print(f"Contextual Embeddings (BERT) Accuracy: {acc_bert * 100:.2f}%")
```

### Observation (Module 3)
* **Comparison:** Compared to Modules 1 and 2, contextual models capture actual word context (polysemy check shows 'bank' changed forms). 
* **Trade-offs:** Contextual word models are significantly computationally heavier than Word2Vec or TF-IDF, but yield richer semantic information.

## Final Word
In practice on larger datasets, Contextualized models (BERT) will routinely out-perform both Frequency-Based and older Prediction-Based methods, at the distinct cost of high computational complexity.
